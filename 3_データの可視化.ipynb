### ヒストグラムの描画
- Survivedに対してヒストグラムを描画し、生存者と死亡者の割合を視覚的に理解します

```python
df["Survived"].plot.hist()
```

### ヒストグラムの装飾
- グラフの大きさやタイトル等、様々な装飾が可能です
- 自分がデータを確認するだけなら上のような表現でよいが、他人に見せる場合は軸ラベルや凡例等がないと不親切ですよね

```python
# グラフサイズの指定 figsize = (横サイズ, 縦サイズ)
df["Survived"].plot.hist(figsize=(2,3))
```

```python
# タイトルもつける title="タイトル"（サイズ指定は上を引継ぎ）
df["Survived"].plot.hist(figsize=(2,3),title="No-survived vs Survived")
```

```python
# グリッドをひいてみる
df["Survived"].plot.hist(figsize=(4,6),title="No-survived vs Survived", grid=True)
```

```python
# x軸とy軸のラベルをつける
# ラベルの設定は前述までのhistの引数による設定ではなく、一度グラフオブジェクトを変数に格納し、set_xlabelおよびset_ylabelで設定する
fig = df["Survived"].plot.hist(figsize=(2,3),title="No-survived vs Survived", grid=True)
fig.set_xlabel("Survived")
fig.set_ylabel("Count")
```

```python
# y=300の位置に赤ラインを引く
plt.axhline(y=300, color="red")
fig = df["Survived"].plot.hist(figsize=(2,3),title="No-survived vs Survived", grid=True)
fig.set_xlabel("Survived")
fig.set_ylabel("Count")
```

```python
# 画像をpng形式で保存する
plt.axhline(y=300, color="red")
fig = df["Survived"].plot.hist(figsize=(2,3),title="No-survived vs Survived", grid=True)
fig.set_xlabel("Survived")
fig.set_ylabel("Count")
plt.savefig("sample_fig.png")
```

### 箱ひげ図の描画
